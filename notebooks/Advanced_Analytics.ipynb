# 1. Historical VaR and CVaR Analysis

Value at Risk (VaR) estimates the maximum expected loss at a given confidence level.

Conditional VaR (CVaR) measures the average loss beyond the VaR threshold.

Confidence Level Used:
- 95%

```python
from scripts.var_cvar_analysis import calculate_var_cvar

var_df = calculate_var_cvar(nav)

var_df.head()

var_df.to_csv(
    "../reports/outputs/var_cvar_report.csv",
    index=False
)
```

### Insight 1

Funds with the most negative VaR values exhibit the highest downside risk and are more vulnerable during market downturns.

# 2. Rolling Sharpe Ratio Analysis

Rolling Sharpe Ratio evaluates risk-adjusted performance over a moving 90-day window.

Formula:

Sharpe Ratio = Mean(Returns) / Std(Returns) × √252

```python
from scripts.rolling_sharpe_analysis import rolling_sharpe

rolling_sharpe(nav)
```

### Insight 2

Funds with consistently positive rolling Sharpe ratios demonstrate stable risk-adjusted performance over time.

# 6. Sector Concentration Analysis (HHI)

Herfindahl-Hirschman Index (HHI) measures portfolio concentration.

Formula:

HHI = Σ(weight²)

Interpretation:
- Higher HHI → More Concentrated Portfolio
- Lower HHI → Better Diversification

```python
holdings = pd.read_csv(
    "../data/raw/09_portfolio_holdings.csv"
)

from scripts.hhi_analysis import sector_hhi

hhi_df = sector_hhi(
    holdings
)

hhi_df.head()
```

### Insight 5

Funds with higher HHI scores are concentrated in fewer sectors and may carry elevated concentration risk compared to diversified funds.
